```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster
import os
```

```python
df = pd.read_csv('Antarctic_Shipwrecks.csv')
```

```python
df.sample(10)
```

```python
df.shape
```

```python
df.isnull().any()
```

```python
m = folium.Map(location=[-54.316667, -36.650000], zoom_start = 3)

marker_cluster = MarkerCluster().add_to(m)

tooltip = "Click for more Info"

for i in range(0,len(df)):
    Mypopup = str("<b>Vessel: " + df.iloc[i]['Vessel'])
    folium.Marker([df.iloc[i]['Lon'], df.iloc[i]['Lat']], popup=Mypopup, tooltip=tooltip).add_to(marker_cluster)
```

```python
m.save(outfile='map.html')
```
